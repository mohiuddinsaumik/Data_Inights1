```python
m = int(input("Enter marks: "))

if m >= 90:
    print("Grade A")
elif 80 <= m <= 89:
    print("Grade B")
elif 70 <= m <= 79:
    print("Grade C")
elif 60 <= m <= 69:
    print("Grade D")
else:
    print("Grade F")
```
